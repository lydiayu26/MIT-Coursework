### Remove outliers and try regression again

```python
data_no_out = data[(data['boxes_per_hr'] < 30) & (data['boxes_per_hr'] > 0)]
```

```python
data_no_out = data_no_out.drop(columns=['Unnamed: 0'])
```

```python
data_no_out
```

```python
plt.figure(figsize=(12,8))
plt.scatter(data_no_out['turn_sum_last_month'], data_no_out['boxes_per_hr'])
plt.xlabel("turnover")
plt.ylabel("productivity")
```

```python
model = sm.OLS(data_no_out['boxes_per_hr'],
               data_no_out[['turn_full_last_month',
                            'turn_part_last_month']]).fit()
```

```python
model.summary()
```

```python
# residual plots for each variable:
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(model, 'turn_full_last_month', fig=fig)
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(model, 'turn_part_last_month', fig=fig)
```

#### Residual plots show that there is a nonlinear relationship between turnover and productivity

remove NaNs in bph_actual_minus_fcast

```python
data_no_out = data_no_out.dropna(subset = ["bph_actual_minus_fcast"])
```

```python
data_no_out['turn_full_sq'] = data_no_out['turn_full_last_month'] * data_no_out['turn_full_last_month']
data_no_out['turn_part_sq'] = data_no_out['turn_part_last_month'] * data_no_out['turn_part_last_month']
```

```python
X1 = data_no_out[['turn_full_last_month',
                'turn_part_last_month',
                  'turn_full_sq',
                  'turn_part_sq']]
#                 'bph_actual_minus_fcast',
#                 'damages_per_sales',
#                 'ft_headcount_wkly',
#                 'leave_behinds',
#                 'missing_wrong',
#                 'ot_hrs',
#                 'pt_headcount_wkly',
#                 'temp_headcount_wkly',
#                 'tot_sales']]
y1 = data_no_out['boxes_per_hr']
```

```python
model2 = sm.OLS(y1, X1).fit()
```

```python
model2.summary()
```

```python

fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(model2, 'turn_full_last_month', fig=fig)
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(model2, 'turn_part_last_month', fig=fig)
```

### correlation matrix to determine features

features that may be good to include:

- pt_headcount_wkly

```python
X = data_no_out[['turn_full_last_month',
                'turn_part_last_month',
                 'turn_full_sq',
                 'turn_part_sq',
                'bph_actual_minus_fcast',
                'damages_per_sales',
                'ft_headcount_wkly',
                'leave_behinds',
                'missing_wrong',
                'ot_hrs',
                'pt_headcount_wkly',
                'temp_headcount_wkly',
                'tot_sales']]
y = data_no_out['boxes_per_hr']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

### multivariate linear regression with selected features

```python
X_train.isnull().sum()
```

```python
m = sm.OLS(y_train, X_train).fit()
```

```python
m.summary()
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(m, 'turn_full_last_month', fig=fig)
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(m, 'turn_part_last_month', fig=fig)
```

#### Evaluate on test data

```python
pred_values = m.predict(X_test)
```

```python
train_pred = m.predict(X_train)
```

```python
rmse_train = np.sqrt(np.mean((y_train - train_pred)**2))
rmse_train
```

```python
rmse = np.sqrt(np.mean((y_test - pred_values)**2))
rmse
```

### Random forest regression

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(X_train, y_train)
```

```python
rfr.score(X_train, y_train)
```

```python
pred_values_rfr = rfr.predict(X_test)
```

```python
mse = np.mean((y_test - pred_values_rfr)**2)
mse
```

```python
rfr.feature_importances_
```

```python
sorted_idx = rfr.feature_importances_.argsort()
plt.barh(X_train.columns[sorted_idx], rfr.feature_importances_[sorted_idx])
plt.xlabel("Random Forest Feature Importance")
```
