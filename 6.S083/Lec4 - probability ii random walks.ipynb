How to know position for random walk

```julia
[jump() for i in 1:10]
```

```julia
function walk(n)
    x = 0 # current position
    for i in 1:n
        x += jump() #update current position
    end
    return x
end
```

```julia
walk(10)
```

```julia
function trajectory(n)
    x = 0 # current position
    xs = [x]  #initialize vector with initial value of x
    
    for i in 1:n
        x += jump() #update current position
    end
    return x
end
```

```julia
v = Int64[] #makes an array of a specific type; much faster
```

```julia
function trajectory(n) #make an array of the different directions/where the walker ends up in after each iteration
    x = 0 # current position
    xs = [x]  #initialize vector with initial value of x, determines element type of array
    
    for i in 1:n
        x += jump() #update current position
        push!(xs, x)
    end
    return xs
end
```

```julia
trajectory(10)
```

## Plotting

```julia
traj = trajectory(50);
```

```julia
using Plots
plot(traj, m=:o)
```

```julia
traj = trajectory(100);
plot(traj, m=:o, size=(500,300))

hline!([0], ls=:dash)
```

```julia
num_walkers = 1000
p = plot(size=(500,400), leg = false)
num_steps = 100

for i in 1:num_walkers
    traj = trajectory(num_steps)
    plot!(traj)
end
p
```

```julia
using Interact
@manipulate for n in 1:20
    traj = trajectory(n)
    plot(traj)
end
```

## Precompute the data

so that it doesn't create a new set of data each time slider increases

```julia
n = 20
traj = trajectory(n)
@manipulate for i in 1:n
    plot(traj[1:i]) #plot up to time i
    xlims!(0,length(traj))
    ylims!(minimum(traj), maximum(traj))
end
```
