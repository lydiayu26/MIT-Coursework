```julia
function euler_SIR(β, γ, S0, I0, R0, h, T)   
    S_traj = [S0]
    I_traj = [I0]
    R_traj = [R0]
    times = [0.0]
    
    S_old = S0
    I_old = I0
    R_old = R0
    
    for i in h:h:T
        dS = -β*S_old*I_old
        dI = β*S_old*I_old - γ*I_old
        dR = γ*I_old
        
        S_new = S_old + h*dS
        I_new = I_old + h*dI
        R_new = R_old + h*dR
        
        push!(S_traj, S_new)
        push!(I_traj, I_new)
        push!(R_traj, R_new)
        push!(times, i)
        
        S_old = S_new
        I_old = I_new
        R_old = R_new
    end
    
    return times, S_traj, I_traj, R_traj
end
```

```julia
function tangent_line(f, a, x)
    m = deriv(f,a)
    y = m*(x-a) + f(a) #point-slope: y-y0 = m(x-x0) where y0=f(a) and x0=a
    
    return y
end
```

```julia
function gradient_descent(f, x0)
    η = 0.01
    steps = 100
    
    curr_x = x0
    for i in 1:steps
        next_x = curr_x - η*deriv(f, curr_x) #we move to the next point in the opposite direction of the derivative and by the step size amount
        
        curr_x = next_x
    end
    
    return curr_x
end
```

A better way to decide when to end the function is to stop is when the different between next_x and curr_x becomes very small.

## 3.3

```julia
function gradient_descent_2d(f, x0, y0)
    η = 0.01
    steps = 500
    
    curr_xy = (x0, y0)
    for i in 1:steps
        g = gradient(f, curr_xy[1], curr_xy[2])
        next_xy = (curr_xy[1] - η*g[1], curr_xy[2] - η*g[2]) #evaluate partial derivative of x and y and move away from that direction
        
        curr_xy = next_xy
    end
    
    return curr_xy
end
```

```julia
abstract type AbstractWalker end
```

```julia
abstract type Abstract2DWalker <: AbstractWalker end
```

```julia
@enum InfectionStatus S I R
```

```julia
struct Location 
    x::Int64
    y::Int64
end
```

```julia
mutable struct Agent <: Abstract2DWalker
    position::Location
    state::InfectionStatus
end
```

```julia
mutable struct Walker2D <: Abstract2DWalker
    position::Location
end
```

```julia
Walker2D(x,y) = Walker2D(Location(x,y))
```

```julia
pos(walker::Abstract2DWalker) = walker.position
```

```julia
function set_pos!(w::Abstract2DWalker, l::Location)
    w.position = l
end
```

```julia
function jump()
    neighbors = (Location(0,1), Location(1,0), Location(-1,0), Location(0,-1))
    return rand(neighbors)
end
```

```julia
function initialize(L, N)
    agents = Agent[] #vector of N agents
    locations = [] #vector of tuples representing locations already occupied
    
    for i in 1:N #create each agent
        x = rand(-L:L) #generate possible location within box
        y = rand(-L:L)
        
        while (x,y) in locations #check to see if this location is already occupied
            x = rand(-L:L)
            y = rand(-L:L)
        end
        
        push!(locations, (x,y)) #add location of new agent to list of ones already occupied
        push!(agents, Agent(Location(x,y), i==1 ? I : S)) #add new agent to vector
    end
    
    return agents
end
```

```julia
function step!(agents, L, pI, pR)
    i = rand(1:length(agents)) #pick a random agent i
    
    #possible new position for i
    nextStep = jump()
    curr_x = pos(agents[i]).x
    curr_y = pos(agents[i]).y
    new_x = curr_x+nextStep.x
    new_y = curr_y+nextStep.y
    while (abs(new_x) > L) || (abs(new_y) > L)
        nextStep = jump()
        new_x = curr_x+nextStep.x
        new_y = curr_y+nextStep.y
    end
    
    newLoc = Location(new_x, new_y)
    
    for j in 1:length(agents) 
        if (newLoc == pos(agents[j])) && (i != j) #see if any other agents currently occupy the proposed new location
            if agents[i].state == I #see if i is infected
                infected = rand() < pI #i infects j with probability pI
                if (agents[j].state == S) && infected #see if j is susceptible and gets infected
                    agents[j].state = I
                end
                
                recovered = rand() < pR #i recovers with probability pR
                if recovered 
                    agents[i].state = R 
                end
                return #exit if i infects j
            end
        end
    end
    
    set_pos!(agents[i], newLoc) #if newLoc not occupied, move i there
    
    recovered = rand() < pR #i recovers with probability pR
    if recovered && (agents[i].state == I)
        agents[i].state = R 
    end
end         
```

```julia
function sweep!(agents, L, pI, pR, N)
    for i in 1:N
        step!(agents, L, pI, pR)
    end
end
```

```julia
function dynamics!(agents, L, pI, pR, N, numSweeps)
    for i in 1:numSweeps
        sweep!(agents, L, pI, pR, N)
    end
end
```

```julia
Len = 20
N = 100
pI = 0.9
pR = 0.01
agents = initialize(Len, N)
Is = []
Ss = []
Rs = []

for i in 1:100
    dynamics!(agents, Len, pI, pR, N, i)
    push!(Is, count(num -> num==I, [agents[i].state for i in 1:N]))
    push!(Ss, count(num -> num==S, [agents[i].state for i in 1:N]))
    push!(Rs, count(num -> num==R, [agents[i].state for i in 1:N]))
end

plot(Ss, label="Susceptible", xlabel = "number of sweeps", ylabel = "mumber of people") 
plot!(Is, label="Infected")
plot!(Rs, label="Recovered")
```

```julia
function L2(β, γ, Ss, Is, Rs)
    T = 100
    SIR_data = euler_SIR(β, γ, 0.99, 0.01, 0.0, 0.1, T)
    SIR_S, SIR_I, SIR_R = SIR_data[2], SIR_data[3], SIR_data[4]
    totalDistance = 0
    
    for i in 1:length(Ss) #(f(x[i], μ, σ) - y[i])^2
        LS = (SIR_S[i] - Ss[i])^2
        LI = (SIR_I[i] - Is[i])^2
        LR = (SIR_R[i] - Rs[i])^2
        totalDistance += LS + 2*LI + LR
    end
    return totalDistance
end
```

```julia
function gradient_descent_2d3(f, x0, y0)
    η = 0.01
    steps = 31
    
    curr_xy = (x0, y0)
    history = [curr_xy]
    
    for i in 1:steps
        g = gradient(f, curr_xy[1], curr_xy[2])
        next_xy = (curr_xy[1] - η*g[1], curr_xy[2] - η*g[2]) #evaluate partial derivative of x and y and move away from that direction
        push!(history, next_xy)
        
        curr_xy = next_xy
    end
    
    return history
end
```
