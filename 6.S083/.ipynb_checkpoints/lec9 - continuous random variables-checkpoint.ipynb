## How to generate numerically exponential random variate

```julia
rand() #uniform distribution in [0,1]
```

```julia
function uniform_pdf(x) # probability that I fall in a particular interval
    if 0 ≤ x ≤ 1
        return 1.0
    else
        return 0.0
    end
end
```

```julia
plot(-2:0.02:2, uniform_pdf, ratio=1)
```

```julia
function uniform_pdf(x, a, b) # probability that I fall in a particular interval
    if a ≤ x ≤ b
        return 1/(b-a)
    else
        return 0.0
    end
end
```

```julia
a = -0.5
b = 1.5

plot(-2:0.02:2, x -> uniform_pdf(x, a, b), ratio=1)
```

```julia
#Generate a random variate which has a uniform dist on [a,b]

uniform(a, b) = a + rand() * (b-a)
```

```julia
scatter([uniform(-0.5, 1.5) for i in 1:1000])
```

```julia
data =[uniform(-0.5, 1.5) for i in 1:100];
```

```julia
histogram(data, nbins=20)
scatter!(data, zero.(data), ms = 2, alpha = 0.5, c=:orange)
```

```julia
data =[uniform(-0.5, 1.5) for i in 1:1000]
histogram(data, nbins=20)
scatter!(data, zero.(data), ms = 2, alpha = 0.5, c=:orange)
```

```julia
data =[uniform(-0.5, 1.5) for i in 1:100000]
histogram(data, nbins=20)
```

```julia
n = 1000
data =[uniform(-0.5, 1.5) for i in 1:n]
plot(sort(data), 1/n:1/n:1) #cumulative distribution function

plot!([-.5, 1.5], [0,1])
```

```julia
F(x) = 1- exp(-λ*x)
```

$y = F(x)$

Solve for $x$ in terms of $y$

```julia
x = rand()

y = -log(x)/λ
```

```julia
xs = rand(10000)

transformation(x, λ) = -log(x) / λ

λ = 2.0
data = transformation.(xs,λ);
```

```julia
scatter(data)
```

```julia
histogram(data, nbins = 100)
```

```julia
xs = rand(100000)

transformation(x, λ) = -log(x) / λ

λ = 2.0
data = transformation.(xs,λ);

stephist(data, nbins=1000, yscale=:log10, normed=true)
plot!(0:0.01:5, x->λ * exp(-λ * x), lw=3)
```
