```julia
#PROBLEM 2c

using Gurobi, JuMP, CSV
data = CSV.read("SelectingHotels.csv", header = true)
profits = data[:, 8]
prices = data[:, 3]
print(prices)
hotelNum = data[:, 1]

model = Model(with_optimizer(Gurobi.Optimizer))
@variable(model, x[1:16], Bin)
@objective(model, Max, sum(x[i]*profits[i] for i in 1:16))
@constraint(model, sum(x[i]*prices[i] for i in 1:16)<=10000000)

optimize!(model)
for i in 1:16
    println("Hotel ", hotelNum[i], " purchased?: ", value(x[i]))
end
println("Total predicted profit: ", objective_value(model))
```

```julia
#PROBLEM 2d
using Gurobi, JuMP, CSV
data = CSV.read("SelectingHotels.csv", header = true)
profits = data[:, 8]
prices = data[:, 3]
hotelNum = data[:, 1]

model = Model(with_optimizer(Gurobi.Optimizer))
@variable(model, x[1:16], Bin)
@objective(model, Max, sum(x[i]*profits[i] for i in 1:16))
@constraint(model, sum(x[i]*prices[i] for i in 1:16)<=10000000)
@constraint(model, sum(x[i] for i in 1:5)<=2)
@constraint(model, sum(x[i] for i in 6:6)<=2)
@constraint(model, sum(x[i] for i in 7:9)<=2)
@constraint(model, sum(x[i] for i in 10:16)<=2)

optimize!(model)
for i in 1:16
    println("Hotel ", hotelNum[i], " purchased?: ", value(x[i]))
end
println("Total predicted profit: ", objective_value(model))
```
