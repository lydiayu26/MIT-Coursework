## Cafe Optimization

Café 53 is famous for their buttery fruit tarts and their fancy decorated sponge cakes. Key ingredients for either cake is their respective dough.

•A pound of butter dough contributes 5 dollars to profit and a pound of sponge dough 6 dollars.

•The bakery can sell enough buttery fruit tarts each day to require 90 lbs of butter dough and enough fancy cake to require 60 pounds of sponge dough.

•Production is mainly constrained by the available mixer time (a pound of butter dough requires 6 minutes of mixing, whereas a pound of sponge dough requires only 4.5 minutes) and the manual labor associated with each type of cake (3 minutes per pound of butter dough, 6 minutes per pound of sponge dough).

•Each day, 420 minutes of labor are available and 465 minutes of Mixer time. 

What are the optimal production quantities to maximize profit?

## Formulation

```julia
using JuMP
using Clp

model=Model(solver=ClpSolver())

#variables
@variable(model, B>=0)
@variable(model, S>=0)

#constraints
@constraint(model, mixer, 6*B+4.5*S<=465)
@constraint(model, labor, 3*B+6*S<=420)
@constraint(model, bdemand, B<=90)
@constraint(model, sdemand, S<=60)

#objective
@objective(model, Max, 6*S+5*B)

#take a look at our model
model
```

```julia
solve(model)
```

```julia
println("B=",getvalue(B))
println("S=",getvalue(S))
println("Profit=",getobjectivevalue(model))
```

# Transportation Optimization

In the high quality bowling ball industry, there are only 4 supplies and 6 primary customers.

Supplier 1, 2, 3, and 4 have a monthly capacity of 100, 200, 300, and 700 bowling balls, respectively.

Customers 1, 2, 3, 4, 5, 6 have monthly demands of 150, 160, 170, 180, 190, and 200 bowling balls, respectively.

The cost to ship these specialty bowling balls is very high and varies with each customer due primarly to location. The further a customer is from the supplier, the more it costs to ship to them.

The cost matrix is below.


What is the cheapest way to meet demand?

## Formulation

```julia
m = 4
n = 6
# Supply
s = [100, 200, 300, 700]
# Demand
d= [150, 160, 170, 180, 190, 200]
# Cost
c = [10 20 30 40 50 60;
     70 80 90 10 20 30;
     40 50 60 70 80 90;
     80 70 60 50 40 30];
```

```julia
using JuMP, Gurobi

model=Model(solver=GurobiSolver())

#variable
@variable(model, x[i=1:m,j=1:n]>=0)

#constraints

#supply
for i=1:m
    @constraint(model,sum(x[i,j] for j=1:n)<=s[i])
end

#demand
for j=1:n
    @constraint(model, sum(x[i,j] for i=1:m)>=d[j])
end

#objective
@objective(model, Min, sum(c[i,j]*x[i,j] for i=1:m, j=1:n))

#look at model
model
```
